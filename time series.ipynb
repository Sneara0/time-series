```python
import pandas as pd
```

```python
# Read the CSV file
df = pd.read_csv("/content/drive/MyDrive/data/dep.csv")

# Remove rows containing zeros
df = df[(df != 0).all(1)]

# Write the modified DataFrame back to a new CSV file
df.to_csv("/content/drive/MyDrive/data/dep_no_zeros.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV files
dep_df = pd.read_csv("/content/drive/MyDrive/data/dep_no_zeros.csv")
non_dep_df = pd.read_csv("/content/drive/MyDrive/data/non_dep_no_zeros.csv")

# Calculate IQR for Activity Levels
dep_iqr = dep_df['Activity (Triaxial-Counts) '].quantile(0.75) - dep_df['Activity (Triaxial-Counts) '].quantile(0.25)
non_dep_iqr = non_dep_df['Activity (Triaxial-Counts)'].quantile(0.75) - non_dep_df['Activity (Triaxial-Counts)'].quantile(0.25)

# Plotting
plt.bar(['Dependent', 'Non-Dependent'], [dep_iqr, non_dep_iqr], color=['blue', 'green'])
plt.ylabel('Interquartile Range (IQR)')
plt.title('Interquartile Range (IQR) of Activity Levels')
plt.show()
```

```python
# Read the CSV files
dep_df = pd.read_csv("/content/drive/MyDrive/data/dep_no_zeros.csv").dropna()
non_dep_df = pd.read_csv("/content/drive/MyDrive/data/non_dep_no_zeros.csv").dropna()

# Calculate IQR for Activity Levels
dep_iqr = dep_df['Activity (Triaxial-Counts)'].quantile(0.75) - dep_df['Activity (Triaxial-Counts)'].quantile(0.25)
non_dep_iqr = non_dep_df['Activity (Triaxial-Counts)'].quantile(0.75) - non_dep_df['Activity (Triaxial-Counts)'].quantile(0.25)

# Plotting
plt.bar(['Dependent', 'Non-Dependent'], [dep_iqr, non_dep_iqr], color=['blue', 'green'])
plt.ylabel('Interquartile Range (IQR)')
plt.title('Interquartile Range (IQR) of Activity Levels without Null Values')
plt.show()
```
